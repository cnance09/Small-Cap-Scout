```python
# Function to split the list into smaller lists

def split_list(tickers, chunk_size):
    return [tickers[i:i + chunk_size] for i in range(0, len(tickers), chunk_size)]

# Split the tickers into lists of up to 2000 each
split_tickers = split_list(tickers, 2000)

split_tickers
```

```python
import yfinance as yf

# Initialize a dictionary to store the results
available_data = []
unavailable_tickers = []

for ticker in tickers:
    try:
        # Fetch data from Yahoo Finance
        stock = yf.Ticker(ticker)

        # Attempt to fetch market cap and shares outstanding
        market_cap = stock.info.get('marketCap')
        shares_outstanding = stock.info.get('sharesOutstanding')

        # If data is retrieved, store it
        if market_cap is not None and shares_outstanding is not None:
            available_data.append({
                'Ticker': ticker,
                'Market Cap': market_cap,
                'Shares Outstanding': shares_outstanding
            })
        else:
            unavailable_tickers.append(ticker)
    except Exception as e:
        # Handle the case where ticker is not found
        unavailable_tickers.append(ticker)

# Convert the available data to a DataFrame
available_df = pd.DataFrame(available_data)

# Save the results
available_df.to_csv('/mnt/data/available_tickers_data.csv', index=False)
pd.DataFrame({'Ticker': unavailable_tickers}).to_csv('/mnt/data/unavailable_tickers.csv', index=False)

# Print the results
print("Data retrieved for available tickers:")
print(available_df)

print("\nTickers not available in Yahoo Finance:")
print(unavailable_tickers)
```
